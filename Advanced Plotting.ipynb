```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(size = 1000)
fig, ax = plt.subplots()

H = ax.hist(x, bins=50, alpha=0.5,
            histtype='stepfilled')
```

```python
fracs = [20,20,20,20,20]
colors = ['b', 'g', 'y', 'r', 'w']
fig, ax = plt.subplots(figsize=(6,6))
pie = ax.pie(fracs, colors=colors, explode=(0,
0,0.05,0, 0), shadow=True,
          labels = ['A', 'B', 'C', 'D', 'E'])
```

```python
fracs = [30, 15, 45, 5, 5]
colors = ['b', 'g', 'r', 'w', 'y']

fig, ax = plt.subplots(figsize=(6, 6))  # make the plot square
pie = ax.pie(fracs, colors=colors, explode=(0, 0, 0.05, 0), shadow=True,
             labels=['A', 'B', 'C', 'D', 'E'])
```

```python
x = np.linspace(0, 10, 30)
dy = 0.1
y = np.random.normal(np.sin(x), dy)

fig, ax = plt.subplots()
plt.errorbar(x,y,dy,fmt='.k')
```

```python
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.fill_between(x, y1, y2, where=(y1 < y2), color="red")
ax.fill_between(x, y1, y2, where=(y1 > y2), color="blue")
```

```python
x = np.random.random(50)
y = np.random.random(50)
c = np.random.random(50)
s = 500 * np.random.random(50)

fig, ax = plt.subplots()
im = ax.scatter(x, y, c=c, s=s, cmap= plt.cm.jet)

fig.colorbar(im, ax=ax)

im.set_clim(0.0, 1.0)
```
