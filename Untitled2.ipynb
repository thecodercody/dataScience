```python
from pylab import *
```

```python
x = np.linspace(0,5,10)
y = x ** 2
```

```python
figure()
plot(x,y,'r')
xlabel('x')
ylabel('y')
title('title')
show()
```

```python
pylab inline
```

```python
subplot(1,2,1)
plot(x,y, 'r--')
```

```python
subplot(1,1,1)
plot(x,y,'r--')
```

```python
subplot(2,2,1)
plot(x,y,'r--')
```

```python
subplot(1,1,1)
plot(x,y,'b--')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```
